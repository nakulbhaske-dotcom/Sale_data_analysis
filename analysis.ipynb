```python
data='''Region,Product,Sales,Profit
North,A,500,120
South,B,300,80
East,C,700,200
West,D,450,150
North,B,600,170
South,C,350,90
East,D,800,250
West,A,400,110'''

print(data)
```

```python
import pandas as pd 
from io import StringIO
```

```python
df=pd.read_csv(StringIO(data))
print(df)
```

```python
df.to_csv('data.csv',index=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("data.csv")

# Show data
print(df.head())

# Basic analysis
print(df.describe())

# Group by region
region_sales = df.groupby("Region")["Sales"].sum()
print(region_sales)

# Plot
region_sales.plot(kind='bar')
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.show()
```
